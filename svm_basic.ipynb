```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn.datasets import make_classification
```

```python
data = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=42)
X, y = data

print(X.shape)
print(y.shape)
```

```python
model = SVC(kernel='linear', C=1.0, degree=2)
model.fit(X,y)
```

```python
# Set style and plot original points
sns.set_style("darkgrid")

# 1. Create a meshgrid of values
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300),
                     np.linspace(y_min, y_max, 300))

# 2. Predict over meshgrid
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# 3. Plot decision boundary
plt.contourf(xx, yy, Z, alpha=0.3, levels=np.linspace(Z.min(), Z.max(), 3), cmap="coolwarm")

# Overlay scatter again for visibility
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, edgecolor="k", s=50)

# Labels and title
plt.title("SVC with Linear Kernel")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.show()
```

# SVM with kernel

```python
model1 = SVC(kernel='rbf', C=1.0, gamma='scale')
model1.fit(X, y)
```

```python
sns.set_style("darkgrid")

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300),
                      np.linspace(y_min, y_max, 300))

z = model1.predict(np.c_[xx.ravel(), yy.ravel()])
z = z.reshape(xx.shape)

plt.contourf(xx, yy, z, alpha=0.3, levels=np.linspace(z.min(), z.max(), 3), cmap="coolwarm")
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, edgecolor="k", s=50)
plt.title("SVC with RBF Kernel")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.show()
```

```python
model2 = SVC(kernel='poly', C=1.0, degree=3)
model2.fit(X, y)
```

```python
sns.set_style("darkgrid")

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300),
                      np.linspace(y_min, y_max, 300))
z = model2.predict(np.c_[xx.ravel(), yy.ravel()])
z = z.reshape(xx.shape)
plt.contourf(xx, yy, z, alpha=0.3, levels=np.linspace(z.min(), z.max(), 3), cmap="coolwarm")
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, edgecolor="k", s=50)
plt.title("SVC with Polynomial Kernel")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.show()
```
