```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# generating a random dataset
X = np.array(range(1000))
X = X*5

# Add random noise
X = X + np.random.normal(100, 500, 1000)
print(X)

Y = np.array(range(1000))
```

```python
# Solving for beta
x_bar = np.mean(X)
y_bar = np.mean(Y)

beta1 = np.sum((X - x_bar) * (Y - y_bar)) / np.sum((X - x_bar)**2)
beta0 = y_bar - beta1 * x_bar

print("Beta0: ", beta0)
print("Beta1: ", beta1)
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(10, 6))

plt.scatter(X, Y, color='blue', label='Data points')
plt.plot(X, beta0 + beta1 * X, color='red', label='Regression line')
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('Target')
plt.legend()
plt.show()
```
