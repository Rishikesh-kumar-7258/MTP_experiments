```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
X, y = make_blobs(n_samples=1000, centers=4, cluster_std=0.60, random_state=0)

plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Synthetic Data for K-Means")
plt.show()
```

```python
kmeans = KMeans()
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='X', label='Centroids')
plt.title("K-Means Clustering on Synthetic Data")
plt.legend()
plt.grid(True)
plt.show()
```

# PCA + K-Means

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D

# Step 1: Generate synthetic 3D data
np.random.seed(42)
cluster_1 = np.random.randn(100, 3) + np.array([5, 5, 5])
cluster_2 = np.random.randn(100, 3) + np.array([-5, -5, -5])
cluster_3 = np.random.randn(100, 3) + np.array([5, -5, 5])
X = np.vstack([cluster_1, cluster_2, cluster_3])

# Step 2: Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

# Step 3: Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Step 4: Plot 3D scatter
fig_3d = plt.figure(figsize=(8, 6))
ax3d = fig_3d.add_subplot(111, projection='3d')
ax3d.scatter(X[:, 0], X[:, 1], X[:, 2], c=labels, cmap='viridis', s=50)
ax3d.set_title("3D View of Clusters")
ax3d.set_xlabel("X")
ax3d.set_ylabel("Y")
ax3d.set_zlabel("Z")
plt.show()

# Step 5: Plot 2D PCA projection
fig_2d = plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', s=50)
plt.title("2D PCA Projection")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.grid(True)
plt.show()
```
